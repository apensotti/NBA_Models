```python
df_matchup_data = t.transform(type='matchup_boxscores')
df_matchup = t.standard_scalar(df_matchup_data)
df_matchup = t.label_encoder(df_matchup,["SEASON_home",'TEAM_ABBREVIATION_home','TEAM_ABBREVIATION_away'])
df_matchup = df_matchup.drop(columns=['TEAM_ID_home','TEAM_NAME_home','GAME_ID','GAME_DATE_home','MATCHUP_home','SEASON_away','TEAM_ID_away','TEAM_NAME_away','GAME_DATE_away','MATCHUP_away','WL_away'])
df_matchup = df_matchup.loc[(df_matchup['TEAM_ABBREVIATION_home'] == 3) & (df_matchup['HOME_GAME_home'] == 1)]
X,y = t.create_windows(df_matchup,window_size=15,type='single',matchup=True)
```

```python
def recent_matchups(df,home,away,n=10):
    home_df = df_matchup.loc[(df_matchup['TEAM_ABBREVIATION_home'] == home) & (df_matchup['HOME_GAME_home'] == 1)].reset_index()
    away_df = df_matchup.loc[(df_matchup['TEAM_ABBREVIATION_home'] == away) | (df_matchup['TEAM_ABBREVIATION_away'] == away)]

    matchup_df = df_matchup.loc[(df_matchup['TEAM_ABBREVIATION_home'] == home) & 
                            (df_matchup['TEAM_ABBREVIATION_away'] == away) & 
                            (df_matchup['HOME_GAME_home'] == 1)].reset_index()

    X = []
    y = []
    for i,row in matchup_df.iterrows():
        row_id = row['GAME_ID']
        X_data = home_df[home_df.loc[home_df['GAME_ID'] == row_id].index[0]-10:home_df.loc[home_df['GAME_ID'] == row_id].index[0]]
        y_data = home_df.loc[home_df.loc[home_df['GAME_ID'] == row_id].index[0]]
        X_data = t.standard_scalar(X_data)
        X_data = t.label_encoder(X_data,['SEASON_home','TEAM_ABBREVIATION_away','TEAM_ABBREVIATION_home'])
        X_data = X_data.drop(columns=['index','TEAM_ID_home','TEAM_NAME_home','GAME_ID','GAME_DATE_home','MATCHUP_home','SEASON_away','TEAM_ID_away','TEAM_NAME_away','GAME_DATE_away','MATCHUP_away','WL_away'])
        X.append(X_data)
        y.append(y_data.loc['WL_home'])

    return np.array(X),np.array(y)
```

```python
from model import ModelConfig
```

```python
model_config = ModelConfig(X,y,[.75,.15,.10])
X_train,y_train,X_test,y_test,X_val,y_val = model_config.train_test_val()
X_train.shape
```

```python
import keras
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.losses import MeanSquaredError,BinaryCrossentropy
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.metrics import RootMeanSquaredError, Accuracy
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score

#,kernel_regularizer=regularizers.L2(0.0001)
model = model_config.create_sequential()
model.add(LSTM(128,'tanh',kernel_regularizer=regularizers.L2(0.0001)))
model.add(Dropout(.2))
model.add(Dense(32,'tanh',kernel_regularizer=regularizers.L2(0.0001)))
model.add(Dropout(.2))
model.add(Dense(64,'tanh',kernel_regularizer=regularizers.L2(0.0001)))
model.add(Dropout(.2))
model.add(Dense(1,'sigmoid'))

model.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=.0001), metrics=[RootMeanSquaredError(),'accuracy'])
```

```python
keras.backend.clear_session()
```

```python
cp = ModelCheckpoint("/home/alexp/NBA_Models/models/models/model5", save_best_only=True,monitor='val_accuracy')
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y_%m_%d-%H_%M_%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)

model.fit(X_train,y_train,validation_data=(X_val,y_val),epochs=250,callbacks=[tensorboard_callback,cp])
```
