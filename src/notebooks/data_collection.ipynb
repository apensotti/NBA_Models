# Team Basic Boxscores

```python
import sqlite3
from sqlite3 import Error
import pandas as pd

def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)

    return conn

conn = create_connection("C:\\Users\\alexp\\src\\NBA_Models\\sqlite\\db\\nba_data.db")
cur = conn.cursor()
cur.execute("SELECT * FROM team_basic_boxscores")
rows = cur.fetchall()

db_list = []

for row in rows:
    db_list.append(row)


basic_boxscores_db = pd.DataFrame(data=db_list)

basic_boxscores_db.columns = ['SEASON', 'TEAM_ID', 'TEAM_ABBREVIATION', 
            'TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'MIN', 'FGM', 'FGA', 
            'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB',
            'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 
            'PLUS_MINUS']

display(basic_boxscores_db)
```
